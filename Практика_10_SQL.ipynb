```python
import sqlite3
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive/')
```

```python
conn = sqlite3.connect('/content/gdrive/My Drive/СибГМУ/Комп Сети/happy_shop.db')

def show(query, connection=conn):
	  return pd.read_sql_query(query,connection)
	 
def table(table_name,connection=conn):
	  return show(f'SELECT * FROM {table_name}',connection)
	 
df_customer_category = pd.read_excel('/content/gdrive/My Drive/СибГМУ/Комп Сети/CustomerCategory_ID.xlsx')
df_customer = pd.read_excel('/content/gdrive/My Drive/СибГМУ/Комп Сети/Customer_table.xlsx')
df_purchase = pd.read_excel('/content/gdrive/My Drive/СибГМУ/Комп Сети/Purchase_table.xlsx')
df_product = pd.read_excel('/content/gdrive/My Drive/СибГМУ/Комп Сети/Product_table.xlsx')
df_purchase_product = pd.read_excel('/content/gdrive/My Drive/СибГМУ/Комп Сети/Purchase_Product_table.xlsx')
df_customer_category.to_sql('customer_category',conn,if_exists='replace',index=False)
df_customer.to_sql('customer',conn,if_exists='replace',index=False)
df_product.to_sql('product',conn,if_exists='replace',index=False)
df_purchase.to_sql('purchase',conn,if_exists='replace',index=False)
df_purchase_product.to_sql('purchase_product',conn,if_exists='replace',index=False)
```

```python
show("""
SELECT strftime('%d-%m-%Y',purchaseDate) AS 'Дата покупки',
customer.name AS 'Имя покупателя',
product.producer AS 'producer', 
product.color AS 'color',
purchase_product.size AS 'size'

FROM ((customer INNER JOIN purchase ON customer.CustomerID = purchase.CustomerID)
INNER JOIN purchase_product ON purchase.purchaseID=purchase_product.purchaseID)
INNER JOIN product ON product.ProductID=purchase_product.ProductID

WHERE product.name = 'Блузка' AND color = 'Белый'
""")
```

```python
show("""
SELECT product.producer AS 'Производитель', SUM(amount*price) AS 'стоимость'

FROM (purchase INNER JOIN purchase_product ON purchase.purchaseID=purchase_product.purchaseID)
INNER JOIN product ON product.ProductID=purchase_product.ProductID
GROUP BY producer
""")
```
